# Display FITS file

Basic display of one FITS file extension.

```python
import os

import numpy as np
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt

from astropy.io import fits

from datapath import DATA
```

```python
data_dirpath = os.path.join(DATA,'astrophotography_data/Andromeda_2022/104MSDCF')
```

```python
image_name = os.path.join(data_dirpath, 'DSC09969.offsets.fits')
extension = 2
```

```python
f = fits.open(image_name)
imarray = f[extension].data
f.close()
```

```python
x0 = 0
x1 = imarray.shape[1]
y0 = 0
y1 = imarray.shape[0]

# x0 = 2000
# x1 = 2500
# y0 = 1100
# y1 = 1600

aslice = np.index_exp[y0:y1, x0:x1]
```

```python
fig = plt.figure(figsize=(9.5,6.5))
# a = fig.add_subplot(1, 2, 1)
plt.imshow(imarray[aslice], vmin=-40, vmax=10)
plt.colorbar()
```
